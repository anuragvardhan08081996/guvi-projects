```python
import pandas as pd
```

```python
csv1 = pd.read_csv("college_1.csv")
csv1.head(5)
```

```python
csv2 = pd.read_csv("college_2.csv")
csv2.head(5)
```

```python
college = pd.concat([csv1,csv2])
college.head(5)
```

```python
df = college[college["CodeKata Score"]>15000] 
df.to_csv("Exceeded expectations.csv")
```

```python
df1 = college[(college["CodeKata Score"]>10000) & (college["CodeKata Score"]<15000)] 
df1.to_csv("Reached_expectations.csv")
```

```python
df2 = college[(college["CodeKata Score"]>7000) & (college["CodeKata Score"]<10000)] 
df2.to_csv("Needs_Improvement.csv")
```

```python
df3 = college[college["CodeKata Score"]<7000] 
df3.to_csv("Unsatisfactory.csv")
```

```python
#Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)
print("Average of previous week geekions:",college["Previous Geekions"].mean())
print("Average of this week geekions:",college["CodeKata Score"].mean())
```

```python
#No of students participated
print("No of students participated:",college["Name"].count())
```

```python
#Average completion of python course or my_sql or python english or computational thinking
print("Average completion of python course:",college["python"].mean())
print("Average completion of mysql course:",college["mysql"].mean())
print("Average completion of python english:",college["python_en"].mean())
print("Average completion of computational_thinking:",college["computational_thinking"].mean())
```

```python
#rising star of the week (top 3 candidate who performed well in that particular week)
#college['total'] = college['python'] + college['mysql']+college['python_en']+college['computational_thinking']
college.iloc[college['CodeKata Score'].argsort()[-3:]]
```

```python
#Shining stars of the week (top 3 candidates who has highest geekions)
college.iloc[college['Previous Geekions'].argsort()[-3:]]
```

```python
#Department wise codekata performence (pie chart)
college.groupby(['Department']).sum().plot(kind='pie', y='CodeKata Score')
```

```python
#Department wise toppers (horizantal bar graph or any visual representations of your choice)
college.groupby(['Department']).max().plot(kind='bar', y='CodeKata Score')
```
